# Part 1

```python
import threading
import time
from pynq.overlays.base import BaseOverlay
import pynq.lib.rgbled as rgbled
base = BaseOverlay("base.bit")
btns = base.btns_gpio
led4 = rgbled.RGBLED(4)
for i in range(4):
    base.leds[i].off()
led4.write(0x0)

def blink(t, d, n):
    '''
      t: number of times to blink the LED
      d: duration (in seconds) for the LED to be on/off
      n: index of the LED (0 to 4)
    '''
    if n <4:
        for i in range(t*2):
            base.leds[n].toggle()
            time.sleep(d)
    elif n == 4:
        for j in range(t):
            led4.write(0x6)
            time.sleep(d)
            led4.write(0x0)
            time.sleep(d)
            
    if n<4:
        base.leds[n].off()
        
def philosopher_t(_f0, _f1, num):
    while True:
        if num == 4:
            led4.write(0x0)
        else:
            base.leds[num].off()
        print("Philosopher-{} is STARVING\n".format(num))
        
        #wait for both forks
        leftFork = _f0.acquire(True)
        rightFork = _f1.acquire(True)
        print("Philosopher-{} is EATING\n".format(num))
        blink(30, 0.1, num) #eat for 3 seconds
        
        _f0.release()
        _f1.release()
        print("Philosopher-{} is NAPPING\n".format(num))
        blink(14, 0.3, num) #nap for 4.2 seconds
        
        if btns[0].read() != 0:
            break
        
    print('Philosopher-{} is done\n'.format(num))
        
philosophers = []
forks = []

#fork locks and array
for i in range(5):
    fork = threading.Lock()
    forks.append(fork)

#philosopher threads
for j in range(5):
    if j < 4:
        p = threading.Thread(target=philosopher_t, args=(forks[j], forks[(j+1)], j))
    elif j == 4:
        p = threading.Thread(target=philosopher_t, args=(forks[j], forks[1], j))
    philosophers.append(p)
    p.start()

for p in philosophers:
    name = p.name
    p.join()
    print('Philosopher {} joined'.format(name))
    
for k in range(4):
    base.leds[k].off()
led4.write(0x0)
print("Terminated\n")
```
